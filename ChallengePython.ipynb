Importando a biblioteca

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Importando as tabelas

```python
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)
```

 Nomeando as tabelas

```python
lojas = [loja1, loja2, loja3, loja4]
nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
nomes_lojas
```

Criando as funções

```python
def calcular_faturamento(df):
    return df['Preço'].sum()

def calcular_avaliacao_media(df):
    return df['Avaliação da compra'].mean()

def calcular_frete_medio(df):
    return df['Frete'].mean()
```

Criando Loops e listas para calcular

```python
faturamentos = []
avaliacoes = []
fretes = []

for loja in lojas:
    faturamentos.append(calcular_faturamento(loja))
    avaliacoes.append(calcular_avaliacao_media(loja))
    fretes.append(calcular_frete_medio(loja))
```

Criando a condicional para verificar a menor desempenho

```python
menor_faturamento = faturamentos[0]
pior_loja = nomes_lojas[0]

for i in range(len(faturamentos)):
    if faturamentos[i] < menor_faturamento:
        menor_faturamento = faturamentos[i]
        pior_loja = nomes_lojas[i]

print("A loja com menor desempenho (faturamento) é:", pior_loja)
```

```python
resumo = pd.DataFrame({
    'Faturamento': faturamentos,
    'Avaliação Média': avaliacoes,
    'Frete Médio': fretes
}, index=nomes_lojas)
```

Gráfico de faturamento

```python
df_faturamento = resumo[['Faturamento']].sort_values('Faturamento')

df_faturamento.plot(
    kind='bar',
    figsize=(14, 10),
    color='tab:orange',
    edgecolor='black'
)
plt.title('Faturamento por Loja')
plt.xlabel('Valor (R$)')
plt.ylabel('Loja')
plt.show()
```

Gráfico de avaliação

```python
df_avaliacao = resumo[['Avaliação Média']].sort_values('Avaliação Média')

df_avaliacao.plot(
    kind='barh',
    figsize=(14, 10),
    color='tab:purple',
    edgecolor='black'
)
plt.title('Avaliação Média por Loja')
plt.xlabel('Nota')
plt.ylabel('Loja')
plt.show()
```

Gráfico Frete médio por loja

```python
df_frete = resumo[['Frete Médio']].sort_values('Frete Médio')

df_frete.plot(
    kind='pie',
    y='Frete Médio',
    figsize=(10, 8),
    autopct='%1.1f%%',
    legend=False
)

plt.title('Distribuição do Frete Médio por Loja')
plt.ylabel('')  # remove o rótulo automático
plt.show()
```
